```python
# nonlocal
```

```python
def a():
    x=10
    def b():
        nonlocal x
        x =20
    b()
    print(x)
a()
```

```python
# How Python search the variable?
# (LEGB rule)
# 1. local
# 2. enclosing
# 3. global
# 4. built-in
```

```python
# builtin namespace
print(dir(__builtins__))
```

```python
country = ['south korea']
def inside_list_function():
    country.append("usa")

inside_list_function()
print(country)
```

```python
country = ['south korea']
def inside_list_function():
    country = ["usa"]

inside_list_function()
print(country)
```

```python
# globals keyword
print(globals())
```
