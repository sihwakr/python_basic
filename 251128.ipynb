## [초미니 프로젝트] 행맨 프로그램 만들기

```python
# hangman
# Here is the word [_,_,_,_,_,_]
# Guess the char?
# [b,_,t,m,a,_]
# [b,a,t,m,a,n]
```

```python
valid_count = 10
final = ['-','-','-','-','-','-']

alphabets = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p',
             'q','r','s','t','u','w','v','x','y','z']

# 1단계: 정답 단어 입력 받기
while True:
    x = input("6자리 영어 문자를 입력하세요: ")

    if len(x) != 6:
        continue

    valid = True
    for alphabet in x:
        if alphabet not in alphabets:
            print("소문자 알파벳만 입력해주세요")
            valid = False
            break
    
    if valid:
        break

list_x = list(x)

# 2단계: 글자 맞추기
while True:
    y = input("한자리 알파벳을 입력해주세요: ")

    if len(y) != 1:
        continue
    
    if y not in alphabets:
        print("소문자 알파벳만 입력해주세요")
        continue

    if y not in list_x:
        print("오답입니다")
        valid_count -= 1
        print("남은 기회:", valid_count)

        if valid_count == 0:
            print("기회를 모두 소진하였습니다. 처음부터 다시 시작해 주세요")
            break
        
        continue

    print("정답입니다")
    
    for i, ch in enumerate(list_x):
        if ch == y:
            final[i] = y
    
    print(final)

    if final == list_x:
        print("완성!")
        break
```
