## Conditional Expression(조건절)에 대해 알아보기

```python
if 10> 20:
    print(True)
else:
    print(False)
```

```python
if 10> 20:
    print(True)
elif 3> 6:
    print(False)
else:
    print("False2")
```

```python
if int(input("how tall are you in cm?"))>180:
    print("You are over 180")
else:
    print("You are less than 180")
```

```python
# Comparison Operator(비교 연산자)
# > Greater than
# >= Greater than or Equal to
# < Less than
# <= Less than or Equal to
# == Equal to
# != Not equal to
```
