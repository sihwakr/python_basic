## [초미니 프로젝트] 강력한 패스워드를 만들어 주는 프로그램을 같이 만들어 보기

```python
lower_alphabets = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']

upper_alphabets = []
for alphabet in lower_alphabets:
  upper_alphabets.append(alphabet.upper())
alphabets = lower_alphabets + upper_alphabets
numbers = [str(i) for i in range(0, 10)]
symbols = ['!', '#', '$', '%', '&', '(', ')', '*', '+']
```

```python
import random
```

```python
strong_password = []
```

```python
alphabet_num = random.randint(0,14)
numbers_num = random.randint(0,9)
symbols_num = random.randint(0,9)
```

```python
while True:
    alphabet_num = random.randint(1,14)
    numbers_num = random.randint(1,9)
    symbols_num = random.randint(1,9)
    
    if alphabet_num+numbers_num+symbols_num ==13:
        print("합이 13입니다")
        print("alphabet_num은", alphabet_num, "입니다")
        print("numbers_num 은",numbers_num,"입니다")
        print("symbols_num 은", symbols_num,"입니다")
        break
```

```python
final =[]
for i in range(alphabet_num):
    alpha = random.choice(alphabets)
    final.append(alpha)

for i in range(numbers_num):
    num = random.choice(numbers)
    final.append(num)

for i in range(symbols_num):
    sym = random.choice(symbols)
    final.append(sym)
```

```python
final
random.shuffle(final)
```
